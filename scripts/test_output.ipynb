```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import cv2   
import sys
import os
```

```python
# img_path = 'UNet/dataset/train/images'
# label_path = 'UNet/dataset/train/targets'
img_path = 'cleaned_images/imgs_teleop_4'
label_path = 'cleaned_labels/imgs_teleop_4'

files_imgs = os.listdir(img_path)
files_labels = os.listdir(label_path)
files = []
for img in files_imgs:
    if img in files_labels:
        files.append(img)
for file in files:
    plt.subplot(121)
    im = cv2.imread(os.path.join(img_path, file))
    im = cv2.resize(im, (64, 64))
    plt.imshow(im)
    plt.axis('off')

    plt.subplot(122)
    label = 127*cv2.imread(os.path.join(label_path, file))
    label = cv2.resize(label, (64, 64))
    plt.imshow(label)
    plt.axis('off')
    plt.show()
```
